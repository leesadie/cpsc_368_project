## Imports

## Loading and cleaning data

### Modality wait times data

```python
# first loading of data - several rows at top and bottom such as title, summary, period that we don't want
wt_df = pd.DataFrame(pd.read_excel('data/Modality_WaitTimes_2020_2021.xlsx', sheet_name=1, header=None))
wt_df
```

```python
# finding the start row of where we want to keep the data and filtering out the irrelevant top rows
start_row = wt_df[wt_df.apply(lambda row: row.astype(str).str.contains('Region', case=False, na=False)).any(axis=1)].index[0]
wt_df_filtered = pd.read_excel('data/Modality_WaitTimes_2020_2021.xlsx', sheet_name=1, skiprows=start_row)
wt_df_filtered
```

```python
# dropping the last 9 rows which contain more notes that we do not want in the dataset
wt_df_filtered = wt_df_filtered.iloc[:-9]
wt_df_filtered
```

```python
# dropping the first column which is an unnamed column that only contains NaN values
wt_df_filtered_final = wt_df_filtered.drop(wt_df_filtered.columns[[0]], axis=1)
wt_df_filtered_final
```

```python
# saving the filtered data as a csv file so it's easier to work with
wt_df_filtered_final.to_csv('data_clean/Modality_WaitTimes_2020_2021_clean.csv')
```

### GDHI data

```python
# dropping the irrelevant first 6 rows
gdhi_df_filtered = gdhi_df.iloc[6:]
gdhi_df_filtered
```

### Workforce data

```python
# lots of irrelevant rows at top and bottom of sheet we will need to get rid of
workforce_df = pd.DataFrame(pd.read_excel('data/Workforce_Stats.xlsx', sheet_name='4. All Staff, NHSE & Org - HC', header=None))
workforce_df
```

```python
# removing the irrelevant first 6 rows that contain notes
workforce_df_filtered = workforce_df.iloc[6:]
workforce_df_filtered
```

```python
# removing the irrelevant last 15 rows that contain notes
workforce_df_filtered = workforce_df_filtered.iloc[:-15]
workforce_df_filtered
```

```python
# dropping the rows that have NaN values for every column - specifically for the rows above that act as spacers in the table
workforce_df_filtered = workforce_df_filtered.dropna(how='all')
workforce_df_filtered
```

```python
# making first row the header of the dataframe
header = workforce_df_filtered.iloc[0]
workforce_df_filtered = workforce_df_filtered[1:]
workforce_df_filtered.columns = header
workforce_df_filtered
```

```python
# renaming the first 4 columns to be more interpretable
new_col_names = ['Region_Num', 'Region_Name', 'Org_Name', 'Org_Code'] + list(workforce_df_filtered.columns[4:])
workforce_df_filtered.columns = new_col_names
workforce_df_filtered
```

```python
# this data starts at 2009 - filtering only for 2020-2021
workforce_df_filtered = workforce_df_filtered[['Region_Num', 'Region_Name', 'Org_Name', 'Org_Code'] + 
                                 [col for col in workforce_df_filtered.columns if str(col).startswith(('2020', '2021'))]]
workforce_df_filtered
```

```python
# we can see that there are many NaN values in between region_nums - because of how the table was organized 
# we are going to fill them in using forward fill to propagate the last valid value downwards

workforce_df_filtered['Region_Num'] = workforce_df_filtered['Region_Num'].ffill()
workforce_df_filtered
```

```python
# let's do the same for region_name
workforce_df_filtered['Region_Name'] = workforce_df_filtered['Region_Name'].ffill()
workforce_df_filtered_final = workforce_df_filtered
workforce_df_filtered_final
```

```python
# saving the final workforce data as csv for easier use
workforce_df_filtered_final.to_csv('data_clean/Workforce_Stats_clean.csv')
```

<div class="alert alert-block alert-info">

**Looking at the counts for each modality with a pie chart**

```python
modality_counts = wt_df['Modality'].value_counts()

plt.figure(figsize=(10, 6))
plt.pie(modality_counts.values, labels=modality_counts.index, autopct='%1.1f%%')

plt.title("Counts of Each Modality", fontsize=14)
plt.show()
```

<div class="alert alert-block alert-info">
<b>Note:</b> MRI seems to be the most common modality in general, followed by radiography, ultrasound, and CT.
</div>

**Over all regions, let's now look at the distribution of wait times for each modality with a boxplot.**

```python
# melting data so months become a single column
wt_df_melted = wt_df.melt(
    id_vars=['Region', 'Org Code', 'Provider name', 'Modality', 'Year'],
    value_vars=['Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 
                'Oct', 'Nov', 'Dec', 'Jan', 'Feb', 'Mar'],
    var_name="Month",
    value_name="Wait Time"
)

# removing NAs for the visualization
wt_df_melted.dropna(subset=["Wait Time"], inplace=True)

# converting months to categorical
month_order = ['Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 
                'Oct', 'Nov', 'Dec', 'Jan', 'Feb', 'Mar']
wt_df_melted["Month"] = pd.Categorical(wt_df_melted["Month"], categories=month_order, ordered=True)

# plot
plt.figure(figsize=(12, 6))
ax = sns.boxplot(x='Modality', y='Wait Time', data=wt_df_melted, palette="Set2")

# reducing number of y-ticks for better readability
ax.yaxis.set_major_locator(mticker.MaxNLocator(integer=True, nbins=8))

plt.title('Distribution of Median Wait Times by Modality')
plt.xlabel('Imaging Modality')
plt.ylabel('Median Wait Time (Days)')
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

<div class="alert alert-block alert-info">
<b>Note:</b> It seems that there are quite a lot of outliers for the majority of imaging modalities, although less so for medical photography. While there are some outliers for PET, there are less compared to the other modalities. Additionally, the interquartile ranges are significantly small for several of the modalities. Some of the modalities seem to have similar medians:  1) CT, fluoroscopy, and PET seem to have similar medians, and 2) MRI, nuclear medicine, and single photon emission CT have similar medians. 1) also seems to be right-skewed, and 2) seems to be fairly symmetric.
</div>

**Let's now take a quick look at the distribution of wait times for each modality in each region, which is what underlies our research question, using faceted box plots.**

```python
g = sns.FacetGrid(wt_df_melted, col="Modality", col_wrap=3, height=4, sharey=True)
g.map_dataframe(sns.boxplot, x="Region", y="Wait Time", hue="Region", palette="Set2")

# reducing number of y-ticks for readability
for ax in g.axes.flat:
    ax.yaxis.set_major_locator(mticker.MaxNLocator(integer=True, nbins=6))

g.set_axis_labels("Region", "Median Wait Time (Days)")
g.fig.suptitle("Distribution of Wait Times by Modality and Region", y=1.05)

plt.show()
```

<div class="alert alert-block alert-info">
<b>Note:</b> It looks like the modality with the greatest variability in IQR for every region is single photon emission CT. Plain radiography and medical photography have little to no variability in IQR for every region, although there are significant outliers in plain radiography. It also seems that for every modality, there are some data that are not mapped to a given region. This may be from the first 9 rows that contain the provider 'England,' which are not mapped to a region, as no other NA or null values were found in the region column. These rows will likely need to be filtered out in further analysis.
</div>

**Let's look at changes over time in the workforce counts for each region.** 

```python
# getting non-date columns only
date_columns = wf_df.columns[4:]

# melting df to get months as one column
wf_df_melted = wf_df.melt(id_vars=["Region_Num", "Region_Name", "Org_Name", "Org_Code"], 
                    value_vars=date_columns, 
                    var_name="Date", value_name="Workforce_Count")

# ensure date column is datetime data type
wf_df_melted["Date"] = pd.to_datetime(wf_df_melted["Date"])

# sort values by region_num and date
wf_df_melted = wf_df_melted.sort_values(by=["Region_Num", "Date"])

# plot workforce count for each region over time
plt.figure(figsize=(12, 6))
sns.lineplot(data=wf_df_melted, x="Date", y="Workforce_Count", hue="Region_Num", marker="o")

plt.xticks(rotation=45)
plt.xlabel("Date")
plt.ylabel("Workforce Count")
plt.title("Workforce Count Over 2020-2021 by Region")
plt.legend(title="Region Number", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.grid(True)

plt.show()
```

<div class="alert alert-block alert-info">
<b>Note:</b> It looks like over time, the workforce counts don't change significantly for every region. What is interesting, however, is that for each region, the workforce counts are nearly linearly stacked for every point. It is thus clear that some regions generally have higher workforce counts than others in general. The Y58 region, South West, seems to have the highest workforce counts in general. This is interesting since South West did not have the highest GDHI; Y56, London, had the highest GDHI and is on the lower end of workforce counts. More analysis is needed to understand why this is and if it would be possible for a model to use this workforce data in combination with the GDHI data to accurately predict wait times for diagnostic imaging services.
</div>
